```python
# print monthly calendar of (year, month)

def isLeapYear(year):
    return year%400==0 or (year%4==0 and year%100!=0)

def monthly_days(year):
    if isLeapYear(year):
        return [31,29,31,30,31,30,31,31,30,31,30,31]
    else:
        return [31,28,31,30,31,30,31,31,30,31,30,31]

def days_passed_in_the_year(year, month, day):
    month = month-1
    # First, get the number of days in the given month
    daysInMonth = monthly_days(year)[month]
    daysBefore = -1
    for iterDay in range(daysInMonth):
        if iterDay >= day:
            pass
        else:
            daysBefore += 1   
    priorMonthDays = 0
    # Check that the month is past January or else you won't be adding days from prior months
    if month > 0:       
        # Then add all days from all prior months of the year
        for monthDays in range(month):
            priorMonthDays += monthly_days(year)[monthDays]       
    # Now add all prior days from previous months to all prior days from this month
    priorDays = priorMonthDays + daysBefore   
    return priorDays   
    

def week_day(year, month, day):   
    weekdays = ['Su', 'Mo', 'Tu', 'We', 'Th', 'Fr', 'Sa']
    # Index value of 'Sa': in this case since python iterates starting at zero (0,1,2,3,4,5,6),
    # weekdayIndex.index('Sa') = 6
    weekdayIndex = weekdays.index('Sa')
   
    if year < 2022:
        daysUntilReferenceDate = 365 - days_passed_in_the_year(year, month, day) 
        # Cycle backwards
        # Note: % gives you the remainder, so you want the number divisible by 7, which is days in a week
        newDay = weekdayIndex - (daysUntilReferenceDate % 7)           
        newDay = weekdays[newDay]       
    else:
        # Cycle forwards
        # Note: % gives you the remainder, so you want the number divisible by 7, which is days in a week
        newDay = (days_passed_in_the_year(year, month, day) % 7) -1           
        newDay = weekdays[newDay]   
    return newDay
        
def print_calendar(year, month):
    days_list = monthly_days(year)
    days = days_list[month-1]
    start_weekday = week_day(year, month, 1)
    n = 8 # number characters per print item
    weekdays = ['Su', 'Mo', 'Tu', 'We', 'Th', 'Fr', 'Sa']
    months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 
              'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
    print('\n')
    print(' '*n, months[month-1], year)
    print(' '*(n-4), '_'*(n*6+4))
    for d in range(0,len(weekdays)):
        print(weekdays[d].rjust(n))
    for d in range(1,32):
        print(str(d).rjust(n))
    
    # finish the rest of printing          <------- I can't figure out the print function yet.
    
def main():
    year, month = eval(input('Year, month = '))
    print_calendar(year, month)
    
if __name__ == '__main__':
    main()
    
main()
```
