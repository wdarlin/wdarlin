```python
import matplotlib.pyplot as plt
xs = [i for i in range(1,80)]
ys = [x*x for x in xs]

fig, ax = plt.subplots()
ax.plot(xs,ys, linewidth =3)
ax.set_title('Square Numbers', fontsize=23)
ax.set_xlabel('Value', fontsize=14)
ax.set_ylabel('Square of Value', fontsize=14)
ax.tick_params(axis='both', labelsize=12)

plt.show()
```

```python
import matplotlib.pyplot as plt
xs = [i for i in range(1,80)]
ys = [x*x for x in xs]

plt.style.use('seaborn')

fig, ax = plt.subplots()
ax.plot(xs,ys, linewidth =3)
ax.set_title('Square Numbers', fontsize=23)
ax.set_xlabel('Value', fontsize=14)
ax.set_ylabel('Square of Value', fontsize=14)
ax.tick_params(axis='both', labelsize=12)

plt.show()
```

```python
ds = [i for i in range(1,5)]
es = [d*d for d in ds]
es2 = [30-e for e in es]

plt.style.use('seaborn')
fig, ax= plt.subplots()
ax.scatter(ds,es2,s=100,color='blue')
ax.plot(ds,es, linewidth=3,color='red')

plt.show()
```
